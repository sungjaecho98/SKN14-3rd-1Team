```python
from keybert import KeyBERT
def test(words):
    doc = " ".join(words)

    candidates = words

    kw_model = KeyBERT()

    keywords = kw_model.extract_keywords(
        doc,
        candidates=candidates,     # ✅ 직접 후보 지정
        use_mmr=False,             # MMR 사용 시 candidates 허용 안됨
        top_n=5
    )

    for keyword, score in keywords:
        print(f"{keyword}: {score:.4f}")

# test(['*본 포장은 안전한 배송을', '위하여 제작되었습니다.(운송용)', '키즈', '건강기능식품450g(15gx30포)', '데일리퍼플', '림', 'netun', '비타민B6,비타민D,비타민E,아연'])
test(['Dongkook', 'nuFit', '마이핏V', '멀티비타 이문', '128', 'MVCIT VMULTVTAMMUNE 128', 'myfit', 'TOTAL 20 VITAMIN& MINERALS', '멀티비타민 및 미네랄', '액상 280m+캡술 7g+정제 8.4g', '[(액상 20m+캡술 500mg+정제600mg)x14개입]'])
```

```python
kw_model = KeyBERT('paraphrase-multilingual-MiniLM-L12-v2')

# 원본 리스트
text_chunks = [
    'Dongkook',
    'nuFit',
    '마이핏V',
    '멀티비타 이문',
    '128',
    'MVCIT VMULTVTAMMUNE 128',
    'myfit',
    'TOTAL 20 VITAMIN& MINERALS',
    '멀티비타민 및 미네랄',
    '액상 280m+캡술 7g+정제 8.4g',
    '[(액상 20m+캡술 500mg+정제600mg)x14개입]'
]

# 전체 문서를 구성 (유사도 기준이 될 문장)
doc = " ".join(text_chunks)

# 키워드 추출
keywords = kw_model.extract_keywords(
    doc,
    candidates=text_chunks,  # 🔑 각 항목을 키워드 후보로 사용
    top_n=5,                 # 상위 N개 추출
    use_mmr=False            # 🔁 MMR 쓰면 candidates 사용 불가
)

# 출력
for keyword, score in keywords:
    print(f"{keyword}: {score:.4f}")
```

```python
from openai import OpenAI
from google.colab import userdata
import os

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')

# ✅ 1. OpenAI 클라이언트 생성
client = OpenAI()  # ← 여기에 본인의 API 키를 입력하세요

# ✅ 2. 비정형 제품 정보 리스트
input_chunks = ['누르세요', '', '대웅제약', '우루사', 'R', '정', '우르소데옥시콜산', '200', '밀리그램', '전문의약품', '간질환', '치료제', '100', '정']
# ✅ 3. 프롬프트 작성
text = ', '.join(input_chunks)

prompt = f"""
다음은 건강기능식품 관련 제품 정보에서 추출한 비정형 텍스트 리스트입니다:

{text}

이 정보를 기반으로 제품을 대표할 수 있는 의미 있는 키워드 5개를 뽑아주세요.
가능하면 아래 기준을 고려하세요:
- 들어온 순서는 중요성과는 무관합니다.
- 제품명으로 생각되는 키워드 위주로 뽑아주세요.
- 한국어로만 된 단어만 뽑아주세요
- 영어로 된 단어는 제외해주세요.
- 유아, 어린이, 임산부, 남성, 여성 등 특정 대상 나이 표현 단어 뽑아주세요.

- '' 안에 있는 것은 하나의 단어로 취급하므로 절대 쪼개지 마세요.


결과는 리스트 형태로 출력해주세요 (예: ['키워드1', '키워드2', ...]).
제품명으로 생각되는 키워드를 앞쪽에 배치해주세요
"""

# ✅ 4. ChatGPT API 호출
response = client.chat.completions.create(
    model="gpt-4o-mini",  # 필요시 "gpt-3.5-turbo"도 사용 가능
    messages=[
        {"role": "user", "content": prompt}
    ],
    temperature=0.3
)

# ✅ 5. 결과 출력
print("📌 추출된 키워드:", response.choices[0].message.content)
```

```python
- 유아, 어린이, 임산부, 남성, 여성 등 특정 대상 나이 표현 단어
- 기능성 또는 성분
- 형태 (젤리/츄잉/정제/캡슐/액상 등)
- 사람의 눈에 자연스러운 단어
- 중복되거나 무의미한 숫자/기호는 제거
-''안에 있는 것은 하나의 단어로 취급하므로 쪼개지 말기
```
